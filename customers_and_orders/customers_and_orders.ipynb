**Question 4** What is the state with the least customers?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')

    header = next(csvreader)
    print("Header:", header)

    states_counts = {}
    for row in csvreader:
        state = row[4].strip().upper()
        if state:
            states_counts[state] = states_counts.get(state, 0) + 1

min_count = min(states_counts.values())

least_states = [s for s, c in states_counts.items() if c == min_count]
print(f"The state(s) with least customers ({min_count}) are: {', '.join(least_states)}")
```

**Question 5:** What is the most common last name?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')

    header = next(csvreader)
    print("Header:", header)

    last_name_counts = {}
    for row in csvreader:
        last_name = row[2].strip().upper()  # columna de apellidos
        if last_name:  # evitar vacíos
            last_name_counts[last_name] = last_name_counts.get(last_name, 0) + 1

# encontrar el apellido más común
most_common_last = max(last_name_counts, key=last_name_counts.get)
print(f"The most common last name is {most_common_last} with {last_name_counts[most_common_last]} customers")
```

```python
# open and evaluate orders.csv

import csv
from collections import defaultdict

# Load the CSV file
with open('/content/sample_data/orders.csv', newline='', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    rows = list(reader)

print(f"Total rows loaded: {len(rows)}")
```

**Question #1: How many unique orders are in the orders.csv file?**

```python
unique_orders = set(row['OrderID'] for row in rows)
print(f"Unique orders: {len(unique_orders)}")
```

**Question #2: What is the average number of items per order (rounded to two decimal places)?**

```python
order_items = defaultdict(int)

for row in rows:
  order_items[row['OrderID']] += 1

avg_items = sum(order_items.values()) / len(order_items)
print(f"Average items per order: { round(avg_items, 2)}")
```

**Question #3: What is the highest number of items per order?**

```python
max_items = max(order_items.values())
max_order_id = [k for k, v in order_items.items() if v == max_items]
print(f"Max items in one order: {max_items} (OrderID: {max_order_id})")
```

**Question #4: What is the number of orders placed in October 2021?**

```python
oct_2021_orders = set()
for row in rows:
    if row['Date'].startswith('2021-10'):
        oct_2021_orders.add(row['OrderID'])
print(f"Orders placed in October 2021: {len(oct_2021_orders)}")
```

**Question #5: Which customer spent the most amount of money in 2021?**
